# Tutorial02: Learn Git

### Basic Operation

1. create a dir:X and initialization

    >$mkdir X

    >$cd X

    >$git init

1. see the status of repository

    >$git status

1. write code in repository

1. add file in temporary storage

    >$ git add filename

    >$ git status

1. commit the file

    >$ git commit -m "message"

    or

    >$ git commit

    then the VIM editor will be opened, and you insert detail message of this change

1. check log
    - >$ git log

    - check log in short 
    >$ git log --pretty=short

    - check log of certain dir/file
    >$ git log dirname/filename

    - check the change
    >$ git log -p filename

1. check the differences 
    - check the differences between work-tree and temporary-storage
    >$git diff

    - check the differences between work-tree and the lastest commit
    >$git diff HEAD

    (it's a good custom that check diff before commit and comfirm)

### Branch Operation
(the effective way for collaborate-coding)

1. check the branch
    >$ git branch

1. create a branch and transfer
    >$ git checkout -b branchname
    
    or
    
    >$ git branch
    
    >$ git checkout branchname
    
1. topic branch and master branch
    >$ git checkout master
    
    >$ git merge --no-ff branchname
    
        then VIM launched, and you enter some message and close it 
    
1. check log as graph
    >$ git log --graph
    

### Change Commit

1. Recall previous version
    >$git reset --hard hard-number(hard-number represents a certain time version)
    
1. Push forward to a future version
    **attention**:
        >$git log
        
        can only check the log before this version
    
    Here we can use git reflog, to check all the opertaion and find the certain hard-number
    
    >$ git reflog
    
    >$git checkout master
    
    >$git reset --hard hard-number

1. solve confliction
    >$git merge --no-ff branchname
    
    >CONFLICT
    
    now,open the conflicted file and change
    
    >$git add
    
    >$git commit
    
1. change the commit message
    >$git commit -amend

1. press the history
    >$git rebase -i HEAD~2
    
    it turns out:
    >pick hard1 message
    
    >pick hard2 message
    
    amend the second line pick->fixup,save the file and close
    it turns out rebase success


### Push to Remote Repository
1. create a repository with the same name

1. add remote repository
    >$git remote add origin git@github.com:username/repositoryname.git
    
    origin is the identification.
    
1. push to remote repository
    * push to master
        >$git push -u origin master
        
    * push to other branch
        >$git checkout branchname 
        
        >$git push -u origin branchname
        

### Pull from Remote

1. clone to desk
    >$git clone git@github.com:username/repositoryname.git
    
    >$git branch -a

1. get branch from remote
    >$git checkout -b branchname origin/branchname

1. edit code

1. push
    >$git push
    
1. get newest version
    >$git pull origin branchname
